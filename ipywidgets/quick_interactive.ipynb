```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact

@interact
def plot(amplitude=(1.0,2.0), frequency=(0.01,2.0), phase=(0,2*np.pi), 
         fun=[('Sin',np.sin),('Cos', np.cos),('Tan',np.tan)],
         pow=False, power=[2,3],
         num=[10,100,200]):

  t = np.linspace(0,2*np.pi,num=num)  
  
  y = amplitude * fun(frequency*2*np.pi*t + phase)

  if pow:
    y = y**power

  plt.plot(t,y)
  plt.ylim(-4,4)
```
