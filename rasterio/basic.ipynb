```python
import numpy as np
import rasterio as rio
from rasterio.plot import show
import matplotlib.pyplot as plt
```

```python
fn = 'gebco_2020_n57.5_s55.0_w10.0_e13.0.tif'   # https://download.gebco.net/
```

```python
ds = rio.open(fn)
```

```python
ds.bounds
```

```python
ds.width, ds.height
```

```python
ds.crs
```

```python
dat = ds.read(1)
```

```python
dat = dat.astype(float)
```

```python
dat[np.where(dat>0)] = np.nan
```

```python
_, ax = plt.subplots(1,1,figsize=(10,10))
show(dat, ax)
```
